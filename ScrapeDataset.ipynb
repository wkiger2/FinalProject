```python
def isNaN(string):
    return string != string
```

```python
def downloadImage(url, file_name): 
    
    #correct user-agent for browser and system -- otherwise server rejects with 406 HTTP Error
    headers = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
    #send get request with user-agent header
    res = requests.get(url, stream = True, headers=headers)

    if res.status_code == 200:
        with open(file_name,'wb') as f:
            shutil.copyfileobj(res.raw, f)
#         print('Image sucessfully Downloaded: ',file_name)
    else:
        print('Image Couldn\'t be retrieved')
```

```python
#insert data into folders named with label
# downloadImage("https://www.dermaamin.com/site/images/clinical-pic/p/photosensitivity/photosensitivity18.jpg", 'test4.jpg')

count = 0

for index, row in dataset.iterrows():
    try: 
        count += 1
        label = row['label']
        label = re.sub(" ", "_", label) #regex get rid of spaces
        url = row['url']

        if isNaN(url):
            print("NAN FOUND")
            continue

        base_dir = r'C:\Users\WilliamKiger\Desktop\DLH_FinalProjectCode\data'
        labeled_dir = os.path.join(base_dir, label)
        #boolean if folder exists
        check_folder = os.path.isdir(labeled_dir)
        
        #call our downloader function
        #if the folder does not exist for label, create one
        if not check_folder:
            os.makedirs(labeled_dir)
            img_name = str(count) + ".jpg"
            path_to_img = os.path.join(labeled_dir, img_name)
            downloadImage(url, path_to_img)
        else:
            img_name = str(count) + ".jpg"
            path_to_img = os.path.join(labeled_dir, img_name)
            downloadImage(url, path_to_img)
            
        #fail safe
        if count >= 17000: 
            break

    except: 
        print('failed at: ' + str(count))
        time.sleep(5)
        continue

print("Finished scraping " + str(count) + " images.")
```
